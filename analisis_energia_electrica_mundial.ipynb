### una vez montado el drive extraigo una hoja especifica del excel de mi interes y la convierto en un dataframe

```python
# obtener la hoja de consumo mundial de energia electrica
# establecer que los encabezados estan situados en la quinta fila
df_enerdata_c = pd.read_excel(archivo_xlsx, sheet_name='Total energy consumption', header=5)
# quitar las columnas y filas ubicadas en extremos y que no se usaran
Total_energy_consumption=df_enerdata_c.iloc[:-4, 1:-2]
# darle un nombre a la primera columna para su identificacion facil
Total_energy_consumption.rename(columns={'Unnamed: 1': 'Paises'}, inplace=True)

# Mostrar las primeras filas del DataFrame
Total_energy_consumption.head()
```

## creare un pequeño codigo que recopile los datos de los paises miembros del brics sin rusia por que no lo tengo y lo reste a el total de los brics

```python
#agregar los datos de rusia restandole a brics los demas paises miembros
paises_brics=['Brazil','India','China','South Africa']

#iniciar el primer dato de rusia
reg2_rusia={'Paises':'Rusia'}

# un for basandome en los años que necesito
for i in range(1990,2024):
  sum_año = 0
  # un segundo for para pasar de pais en pais
  for j in paises_brics:
    v_actual = Total_energy_consumption.loc[Total_energy_consumption['Paises'] == j , i ].values[0]
    sum_año +=v_actual
  #una vez obtenido la suma total de un año de los diferentes paises restarlo a los brics
  v_rusia = Total_energy_consumption.loc[Total_energy_consumption['Paises'] == 'BRICS' , i ].values[0] - sum_año
  #agregar el dato de rusia al diccionario
  reg2_rusia[i] = v_rusia

#imprimir el diccionario para comprobacion
print(reg2_rusia)

#agregar el diccionario a el dataset
Total_energy_consumption.loc[len(Total_energy_consumption)] = reg2_rusia

#confirmar que ya esta en el data set
Total_energy_consumption.head()
```

## Obteniendo datos de geometria de los paises

```python
# carga la data de la geometria de los paises
prueba = px.data.gapminder()
prueba2 = prueba.loc[prueba.groupby('country')['year'].idxmax()]
prueba2.drop(['year','lifeExp','pop','gdpPercap'], axis = 1, inplace= True)
reg_rusia = {'country': 'Rusia', 'continent': 'Europe', 'iso_alpha': 'RUS', 'iso_num': 643}
prueba2.loc[len(prueba2)] = reg_rusia
prueba2.head()
```

## pequeño analisis para determinar cual dataset es mas completo

```python
paises_df1 = set(Total_energy_consumption['Paises'])
paises_df2 = set(prueba2['country'])

energy_consumption_mundial = paises_df1 - paises_df2  # Países en df1 pero no en df2
geopandas_world = paises_df2 - paises_df1  # Países en df2 pero no en df1
# Convertir a listas
paises_unicos_df1 = list(energy_consumption_mundial)
paises_unicos_df2 = list(geopandas_world)
#acomodar las listas en orden alfabetico
paises_unicos_df1.sort()
paises_unicos_df2.sort()
print("Países únicos en energy_consumption_mundial que no están en geopandas_world:", len(paises_unicos_df1))
print("Países únicos en geopandas_world que no están en energy_consumption_mundial:", len(paises_unicos_df2))
```

## debido al anterior analisis se conbinan ambos dataset pero se establece como principal el de gapminder

```python

#combinando con el data set de produccion de energia electrica
cons_combinado = pd.merge(Total_energy_consumption, prueba2, left_on='Paises', right_on='country', how='right')

# eliminar la primera columna 'paises' (para luego reemplazarla con la columna 'name' )
cons_combinado.drop('Paises', axis=1, inplace=True)

# mover la columna 'country' a la primera posicion para una mejor comprension
cons_combinado = cons_combinado[['country'] + [col for col in cons_combinado.columns if col != 'country']]

# cambiar el nombre de la columna 'name' por 'Paises'
cons_combinado.rename(columns={'country': 'Paises'}, inplace=True)

# mostrar las primeras filas del resultado
cons_combinado
```

## ya con los datos preparados implemento un grafico en DASH que contenga una vista de los paises sobre un mapa mundial y que su tasa de consumo basada en un año se muestre dependiendo de un slider en la parte superior ademas que los paises que mas consumen mostrandose a la derecha

```python
import dash
import pandas as pd
import numpy as np
from dash import dcc
from dash import html
import plotly.express as px
import plotly.graph_objs as go

# Crear la aplicación Dash
app = dash.Dash(__name__)

# Datos para el mapa mundi
data = cons_combinado.copy()

# lista de paises para el drop down
resul_p = list(set(data['Paises']) - set(paises_unicos_df2))
resul_p.sort()


# Layout de la aplicación
app.layout = html.Div([
    html.H1("Mapa Mundi Interactivo"),

    html.Div([
        dcc.Graph(id='mapa-mundi')
    ], style={'width': '59%', 'display': 'inline-block'}),

    html.Div([dcc.Graph(id='bar-group')
    ], style={'width': '39%', 'display': 'inline-block'}),

    html.Div([html.Label("Año:"),
        ], style={'width': '100%', 'display': 'inline-block'}),

    html.Div([dcc.Slider(id='year-slider', min=1990, max=2023,
                  value=2012,  # Año inicial
                  marks={str(year): str(year) for year in range(1990, 2024)},
                  step=None ),]),
    html.Div([
        dcc.Graph(id='line-country')
        ]),

    html.Div([
        html.Div([
            "año que desea predecir: ",
            dcc.Input(id='fanio-input', value=2020, type='number')
        ], style={'width': '30%', 'display': 'inline-block', 'text-align': 'center'}),

        html.Div([
            dcc.Dropdown(
                id='pais-dropdown',
                options=[{'label': pais, 'value': pais} for pais in resul_p],
                value=resul_p[0]  # Valor inicial del dropdown
            ),
        ], style={'width': '30%', 'display': 'inline-block', 'text-align': 'center'})
    ], style={'display': 'flex', 'justify-content': 'center', 'align-items': 'center'}),

    html.Div([dcc.Graph(
        id='modelo-graph')
    ]),

],style={'background':'white'})

@app.callback(
    dash.dependencies.Output('mapa-mundi', 'figure'),
    dash.dependencies.Output('bar-group', 'figure'),
    dash.dependencies.Output('line-country', 'figure'),
    dash.dependencies.Output('modelo-graph', 'figure'),
    [dash.dependencies.Input('year-slider', 'value'),
     dash.dependencies.Input('pais-dropdown', 'value'),
     dash.dependencies.Input('fanio-input', 'value')]
)

def update_map(selected_year, selected_pais, selected_fanio):
  #capturando los datos que se van a mostrar
  year = selected_year
  pais = selected_pais


  try:
    f_year = int(selected_fanio)
  except ValueError:
    f_year = 2020

  try:
    data_ord = data.sort_values(by=year, ascending=False)
    #armando el grafico principal del mapa mundi
    fig = px.choropleth(data_ord,
                          locations="iso_alpha",
                          color='Paises',
                          hover_name=year,
                          color_discrete_sequence=px.colors.qualitative.Prism,
                          title=f'consumo de energia en {year} por paises')
    fig.update_layout(
          showlegend=False,
          xaxis=dict(fixedrange=True),  # Bloquear zoom en el eje x
          yaxis=dict(fixedrange=True),  # Bloquear zoom en el eje y
          dragmode=False  # Desactivar el modo de arrastre
      )
    #estableciendo una lista con los 10 paises con mayo produccion
    top_10 = data.nlargest(10, year)

    fig2 = px.bar(top_10,
                  x=year,
                  y='Paises',
                  color='Paises',
                  color_discrete_sequence=px.colors.qualitative.Prism,
                  title=f'10 paises con mayor consumo de energia en {year}')
    fig2.update_layout(showlegend=False)

    # aqui se arma el grafico de lineas de los 10 paises mostrados
    df_long = pd.melt(top_10, id_vars=['Paises'], var_name='Año', value_name='Mtoe')
    fig3 = px.line(df_long, x='Año', y='Mtoe', color='Paises', markers=True, log_y=True,
                title='Evolución de datos por país (1990-2023)')
    fig3.update_layout(
        yaxis=dict(
            showticklabels=False  # Ocultar etiquetas del eje y
        )
    )

    #aqui se mostrar el grafico del modelo predictivo
    datos_pais = cons_combinado.loc[data['Paises'] == pais].values[0]
    datos_valores = datos_pais[1:-3]
    datos_anio = np.arange(1990, 2024).astype(float)
    x = datos_anio.reshape(-1, 1)
    y = datos_valores.astype(float)

    coef = np.polyfit(datos_anio, y, 5)
    y_poly_pred = np.polyval(coef, datos_anio)
    prediccion = np.polyval(coef, f_year)



    # Crear trazas para el gráfico
    trace1 = go.Scatter(
          x=datos_anio,
          y=y,
          mode='lines+markers',
          name='Datos originales'
      )

    trace2 = go.Scatter(
          x=datos_anio,
          y=y_poly_pred,
          mode='lines',
          name='Modelo Polinómico',
          line=dict(color='green')
      )

    # Crear layout del gráfico
    layout = go.Layout(
          title=f'La predicción aproximada en Mtoe para el año {f_year} es de {prediccion} para {pais}.',
          xaxis=dict(title='Año', showgrid=True),
          yaxis=dict(title='Mtoe', showgrid=True),
          showlegend=True
      )

    # Crear figura con las trazas y el layout
    fig4 = go.Figure(data=[trace1, trace2], layout=layout)
    # Devolver las figura para que se actualice el gráfico
    return fig, fig2, fig3, fig4
  except Exception as e:
    print(f"Error al generar las figuras: {e}")
    return dash.no_update, dash.no_update, dash.no_update, dash.no_update


# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=True)
```
